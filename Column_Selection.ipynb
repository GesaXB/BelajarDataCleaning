 Pemilihan kolom (columns selection) pada Pandas Data Frame berdasarkan tipe data


```python
# import moduls
import pandas as pd
import numpy as np

# cek versi
print(pd.__version__)
print(np.__version__)
```

```python
# membuat DataFrame
n_rows = 5
n_cols = 2
cols = ['bil_pecahan', 'bil_bulat']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns = cols)
df['bil_pecahan'] = df['bil_pecahan'].astype('float')

df.index = pd.date_range('2023-01-01', periods=n_rows, freq='H')
df = df.reset_index()

df['teks'] = list('ABCDE')

df
```

Memilih kolom dengan tipedata numerik

```python
df.select_dtypes(include= 'number') # menampilkan kolom bil_pecahan dan bil_bulat
df.select_dtypes(include= 'float') # menampilkan kolom bil_pecahan
df.select_dtypes(include= 'int') # menampilkan kolom bil_bulat
```

Memilih kolom dengan tipedata object atau string

```python
df.select_dtypes(include= 'object')
```

Memilih kolom dengan tipedata object atau string

```python
df.select_dtypes(include= 'datetime')
```

memilih kolom dengan kombinasi tipedata

```python
df.select_dtypes(include= ['number', 'object'])
```
