```python
# import pandas dan numpy
import pandas as pd
import numpy as np
```

```python
# membuat dataframe dengan random number (numpy)

n_rows = 5
n_cols  = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)),
                  columns = cols)
df
```

```python
# prefix
df.add_prefix("Kolom_")
```

```python
# suffix
df.add_suffix("_field")
```

```python
df
```

```python
# rows selection or
df[(df["A"] == 1) | (df["A"] == 3)]
```

```python
# rows selection isin
df[df["A"].isin([1,3])]
```

```python
# rows selection not
df[~df["A"].isin([1,3])]
```

```python
# membuat dataframe
data = ({'COl 1' : ['1','2','3',"Teks"],
        'COl 2' : ['1','2','3','4']})
df = pd.DataFrame(data)
df.dtypes # mengetahui tipedata
```

```python
# konversi tipe data
df_x = df.astype({'COl 2' : 'int'})
df_x.dtypes
```

```python
# konversi semua tipe data (to_numeric)
df_y = df.apply(pd.to_numeric, errors='coerce')
df_y
```
