```python
# import moduls
import pandas as pd
import numpy as np
```

```python
# cek Version
print(pd.__version__)
print(np.__version__)
```

```python
# Membuat DataFrame
n_rows = 10
n_cols = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)),
                  columns = cols)
df
```

Selection dengan operator logika | (or)

```python
df[(df["A"] == 4) | (df["A"] == 2)]
```

Selection dengan fungsi isin

```python
df[df["A"].isin([4,2])]
```

Menggunakan Operator ~

untuk menampilkan colom baris yang nilainya tidak sama dengan parameter yang di cari

```python
df[~df["A"].isin([4,2])]
```
