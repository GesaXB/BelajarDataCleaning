1) Menyertakan prefik dan suffix pada semua kolom DataFrame

```python
# import moduls
import pandas as pd
import numpy as np
```

```python
# Cek Version
print(pd.__version__)
print(np.__version__)
```

```python
# Membuat DataFrame
n_rows = 5
n_cols = 4
cols = tuple("HALO")

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows,n_cols)),
                 columns= cols)
df
```

 Menyertakan Prefix

String di depan sebuah string tertentu

```python
df.add_prefix("Kolom_")
```

Menyertakan Suffix

String di belakang string

```python
df.add_suffix("_field")
```
